```python
# 印出1-10, 兩種印法
for i in range(30):   # 0~9
  print(i+1 ,  ","  , end="")   # end="" 不空行
print()  #控行

for i in range(10):
  print("1725367125367156731573",end="")
  print(i+1)   # 空行
```

```python
# 請印出 1~100 3 的倍數，不斷行
for i in range(101):
  if( i%3 ==0 ) :    # module 取餘數
    print(i,",",end="")

# 請印出 1~100 3 and 5 的倍數，不斷行
# 邏輯判斷 A and B ---> A,B同時成立 , A or B --> A或B 其中一個成立  ,not A --> A不成立
print()
print("印出 1~100 3 and 5 的倍數--->",end="")
for i in range(100):
 if( i%3==0 and i%5==0): 
   print(i,",",end="")
```

```python
# 印出 100 個星星,不斷行
for i in range(100): 
  print("*",end="")
print()

# 印出 10x10 的星星方塊
for i in range(10):
  print(i,end="")
  for j in range(10):
    print("* ",end="")
  print()  

for i in range(10):
  print(i,end="")
  for j in range(10):
    print(j," ",end="")
  print() 
```

```python
# 巢狀迴圈
for i in range(1,6):
  for j in range(1,6):
    print(i,",",j)
```

```python
# 印出三角形
for i  in range(10):   # 0-9
  print(i,",",end="")
  for j in range(i):  # <--- Range(i)
    print("* * ",end="")
  print()   #斷行
  
print()
for i  in range(10,0,-1):   # 10~0
  print(i," ",end="")
  for j in range(i): print("# # ",end="")
  print()

```

```python
for i  in range(10):
  print(i,end="")
  for j in range(20-i): print(" ",end="")  # key <----- 插入空白
  for j in range(i): print("*",end="")
  print()

  
# python特有的寫法
for i  in range(10): print(" "*(12-i) + "*"*i )
```
