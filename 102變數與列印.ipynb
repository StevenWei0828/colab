integer : 帶正負號的 32 位元(bit) (4 位元組 byte) 整數，值的範圍從 -2,147,483,648 到 2,147,483,647。

float : 帶正負號的 IEEE 64 位 (8 個位元組) 雙精確度浮點數，其範圍介於-1.79769313486231570 E + 308 到-4.94065645841246544 E-324 之間的負數值，以及 4.94065645841246544 E-324 到 1.79769313486231570 E + 308 的正值。 雙精度浮點數會儲存實數的近似值。

string : 不帶正負號的16位 (2 個位元組) 程式碼片段的順序，範圍介於0到65535之間。 每個程式 代碼點 或字元碼都代表一個 Unicode 字元。 字串可包含0到大約 2000000000 (2 ^ 31) Unicode 字元。

Boolean : 保留可以是或的值 True False 。 關鍵字 True ，並 False 對應至變數的兩種狀態 Boolean 

```python
#不用宣告變數，直接給內容，依照內容定型別
num1 = 10   #integer 整數，沒有小數點
num2 = 3.1416 #float  浮點數
flag = True  #Boolean : 布林 True / False
str1 = "This a book"   #string 字串
str2 = "123"  #sring
```

print(%d,%f,%s) 格式化輸出命令

格式 print("%d格式化命令" % (變數列) )

%5d 列印整數，占5個字元

%5s 列印字串，占5個字元

%8.2f 列印浮點數，共占8個字元，2位數小數點

```python
name="Steven"
score=80.6314
print("%5s的分數，用整數5d列印--->%5d" % (name, score))
print("%5s的分數，用浮點數8.2f列印---->%8.2f" % (name, score))

```

```python
##最重要功能－－－對齊
a=34
b=8987
c=344
d=1
print("沒有對齊-->")
print(a,a)
print(b,b)
print(c,c)
print(d,d)
print("有對齊-->")
print("%5d%5d"%(a,a))
print("%5d%5d"%(b,b))
print("%5d%5d"%(c,c))
print("%5d%5d"%(d,d))
```
